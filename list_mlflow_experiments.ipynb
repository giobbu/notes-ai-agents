```python
from mlflow.tracking import MlflowClient

DATABASE_URI = "sqlite:///mlflow.db"  # point to your database

client = MlflowClient(tracking_uri=DATABASE_URI)
experiment_name = "Ollama"

# List deleted experiments
deleted_exps = client.search_experiments(view_type=2)  # view_type=2 → deleted
deleted_exps
```

```python
# Restore
for exp in deleted_exps:
    if exp.name == experiment_name:
        client.restore_experiment(exp.experiment_id)
        print(f"Restored experiment {exp.name} (ID: {exp.experiment_id})")
```
