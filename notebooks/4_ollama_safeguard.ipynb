```python
import ollama
from loguru import logger
```

## Load safeguard model

```python
safeguard_llm_model = "gpt-oss-safeguard:20b"
```

## Input Guard

* ### Define the policy
* ### Add policy to the system content
* ### Check user prompt

```python
POLICY = """
You are a safety classifier.

Policy:
- Block instructions involving violence, illegal activity, or self-harm.
- Allow normal questions and programming help.

Return JSON:
{
  "allowed": true/false,
  "reason": "short explanation"
}
"""
```

```python
def check_user_prompt(prompt):
    """
    Check if the user prompt is safe according to the policy.
    
    Args:
        prompt (str): The user input to check.
    Returns:
        str: The JSON response from the safeguard model.
    """
    response = ollama.chat(
        model=safeguard_llm_model,
        messages=[
            {"role": "system", "content": POLICY},
            {"role": "user", "content": prompt}
        ]
    )
    return response["message"]["content"]
```

```python
user_prompt = "How can I hack a wifi network?"

verdict = check_user_prompt(user_prompt)

logger.info("Safeguard result:", verdict)
```

```python
llm_model= "gpt-oss:20b"
```

```python
# Based on the safeguard verdict, decide whether to proceed with the main LLM
if '"allowed": true' in verdict:
    answer = ollama.chat(
        model=llm_model,
        messages=[{"role": "user", "content": user_prompt}]
    )
    logger.info(answer["message"]["content"])
else:
    logger.warning("Request blocked by safeguard")

```
