```python
import ollama
from loguru import logger
```

```python
llm_model = "gpt-oss:20b"
```

```python
class Agent:
    
    def __init__(self, system=""):
        self.system = system
        self.messages = []
        self.tools_used = []
        if self.system:
            self.messages.append({"role": "system", "content": system})

    def __call__(self, message):
        try:
            self.messages.append({"role": "user", "content": message})
            result = self.execute()
            self.messages.append({"role": "assistant", "content": result})
            return result
        except Exception as e:
            logger.error(f"Error during agent call: {e}")
            return "An error occurred while processing your request."

    def execute(self):
        try:
            response = ollama.chat(
                    model=llm_model,
                    messages=self.messages,
                    options={"temperature": 0.5},
                )
            return response["message"]["content"]
        except Exception as e:
            logger.error(f"Error during agent execution: {e}")
            return "An error occurred while processing your request."
```

```python
assistant_agent = Agent(system="You are a helpful assistant. You always provide accurate and concise answers to user queries. If you don't know the answer, say you don't know. If don't understand the question, ask for clarification. Provide only the answer, without any additional commentary or explanation.")
```

```python
student_agent = Agent(system="You are a curious student. You ask questions to learn more about the world. You can ask follow-up questions to clarify your understanding. Always be polite and respectful in your questions. At the end of your conversation, summarize what you have learned in a concise manner.")
```

```python
max_interactions = 2
student_message = "Could you help me understand the most ecological means of transportation?"

for i in range(max_interactions):
    assistant_message = assistant_agent(student_message)
    student_message = student_agent(assistant_message)

final_message = "Could you summarize what you have learned about ecological transportation?"
summary = student_agent(final_message)

print("Summary of what the student learned:")
print(summary)
```
